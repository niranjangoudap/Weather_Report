```python
import numpy as np
import matplotlib.pyplot as plt
Days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
max_temp = [35, 26, 31, 30, 28, 25, 29]
min_temp = [25, 23, 26, 25, 27, 23, 21]
wind = [13, 17, 14, 21, 28, 18, 20]
ypos = np.arange(len(Days))
plt.bar(ypos - 0.2, max_temp, color="tomato", width=0.2, label='Max Temp (°C)')
plt.bar(ypos, min_temp, color="gold", width=0.2, label='Min Temp (°C)')
plt.bar(ypos + 0.2, wind, color="lightgreen", width=0.2, label='Wind (km/hr)')
for i in range(len(Days)):
    plt.text(ypos[i] - 0.2, max_temp[i] + 0.5, str(max_temp[i]), ha='center', va='bottom', fontsize=9, color='black')
    plt.text(ypos[i], min_temp[i] + 0.5, str(min_temp[i]), ha='center', va='bottom', fontsize=9, color='black')
    plt.text(ypos[i] + 0.2, wind[i] + 0.5, str(wind[i]), ha='center', va='bottom', fontsize=9, color='black')
plt.xticks(ypos, Days)
plt.xlabel("Days", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.title("Temperature and Wind Variation Over a Week", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
